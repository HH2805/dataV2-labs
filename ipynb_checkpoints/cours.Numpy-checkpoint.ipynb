```python
c=np.random.random((2,3,2,4))
c
```

```python
print(c[0]), print()
print(c[0,0]), print()
print(c[0,0,1]), print()
print(c[0,0,:,2]), print()
print(c[0,0,1,2])
```

```python
    # Opérations:
# sum of each column in matrix c:
print(np.sum(c, axis=0)), print(), print()
# sum of each row in matrix c:
print(np.sum(c, axis=1)), print(), print()
print(np.sum(c[0])),print()
print(np.sum(c[0,0])),print()
print(np.sum(c[0,0,1])),print()
print(np.sum(c[0,0,:,2]))
```

```python
    # de même :
print(np.mean(c[0,0,1])),print(),
print(np.mean(c[0,0,:,2]))
```
