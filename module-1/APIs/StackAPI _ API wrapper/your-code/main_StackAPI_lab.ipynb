# StackAPI (Python API wrapper for the Stack applications)

#### Import the necessary libraries here:

```python
# your code here
from pprint import pprint
import pandas as pd
```

#### Question 1: Find the questions and answers of last month on the Stack applications.

```python
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
```

```python
questiondict = SITE.fetch('questions', fromdate="2021-02-01", todate="2021-02-28")
type(questiondict)
```

```python
"""The StackAPI doc mentions that "Returned data will appear 
in the items key of the resulting dictionary"."""
print(type(questiondict['items']))
print(len(questiondict['items']))
pprint(questiondict['items'][1])
```

```python
pd.DataFrame(questiondict['items'])
```

#### Question 2: Find the most voted question today with at least a score of 3 and tagged with 'python'. 

```python
# your code here
pythonquest = SITE.fetch('questions', fromdate='2021-03-27 00:00:01', todate="2021-03-28 00:00:00",
    min=3, tagged="python", sort='votes')
pd.DataFrame(pythonquest['items'])
```
