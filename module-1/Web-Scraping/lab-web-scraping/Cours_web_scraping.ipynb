```python
import requests
import pandas as pd
```

```python
url_lm = 'https://www.lemonde.fr/economie/article/2020/07/22/delphine-ernotte-reconduite-a-la-tete-de-france-televisions_6046981_3234.html'
html_lm = requests.get(url_lm).content
```

```python
html_lm[0:1000]
```

```python
from bs4 import BeautifulSoup
```

```python
soup_lm = BeautifulSoup(html_lm, 'html')
print(type(soup_lm))
print(len(soup_lm))
soup_lm
```

```python
tags = ['h1', 'h2', 'h3', 'h4', 'h6', 'h7', 'p']
soupsel = soup_lm.find_all(tags)
```

```python
text = [element.text for element in soupsel]
print(type(text))
print(set([type(element) for element in text]))
print(text)
```

```python
print([''.join(e.split('\xa0')) for e in text])
```

```python
url_le = 'https://en.wikipedia.org/wiki/List_of_European_countries_by_life_expectancy'
html_le = requests.get(url_le).content
soup_le = BeautifulSoup(html_le, "html")
```

```python
table = soup_le.find_all('table', {'class': 'sortable wikitable'})[0]
table
```

```python
for x in [row.text for row in table.find_all('tr')][1:]:
    print(x)
    break
```

```python
rows = table.find_all('tr')
res = []
for rowx in [row.text.strip().split("\n") for row in rows]:
    rowx = [elt.strip() for elt in rowx if elt != ''][1:]
    res.append(rowx)
print(res)
```

```python
colnames = res[0]
data = res[1:]
df = pd.DataFrame(data, columns=colnames)
df
```

```python
res = requests.get("http://books.toscrape.com/")
soup_b = BeautifulSoup(res.content, 'html')
```

```python
soup_b.prettify
```

```python
res = []
for book in soup_b.find_all('li', {'class': 'col-xs-6 col-sm-4 col-md-3 col-lg-3'}):
    title = book.h3.a['title']
    div_prodprice = book.find('div', {'class': 'product_price'})
    price = div_prodprice.find_all('p')[0].text
    stock = div_prodprice.find_all('p')[1].text.strip()
    res.append([title, price, stock])
res_df = pd.DataFrame(res)
```
