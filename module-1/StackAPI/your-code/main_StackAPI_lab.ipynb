# StackAPI (Python API wrapper for the Stack applications)

#### Import the necessary libraries here:

```python
# your code here
import pandas as pd
from stackapi import StackAPI

SITE = StackAPI('stackoverflow')
```

#### Question 1: Find the questions and answers of last month on the Stack applications.

```python
SITE = StackAPI('stackoverflow')
```

```python
questions = SITE.fetch('questions', fromdate="2021-02-01", todate="2021-02-28")
```

```python
# StackAPI mentions that "Returned data will appear in the items key of the resulting dictionary.
questions['items'][1]
```

```python
pd.DataFrame(questions['items'])
```

#### Question 2: Find the most voted question today with at least a score of 3 and tagged with 'python'. 

```python
# your code here
questions = SITE.fetch('questions', fromdate='2021-03-27 00:00:01', todate="2021-03-28 00:00:00",
    min=3, tagged="python", sort='votes')
pd.DataFrame(questions['items'])
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
# your code here
answers = SITE.fetch('answers', ids=[6784, 6473])
answers['items']
```
