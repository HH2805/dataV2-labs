### Create a new column

```python
df['My New Column'] = 'Please complete'
df.head()
```

```python
# Copy a column
df['Duplicated Column'] = df['Rotten Tomatoes']
df.head()
```

```python
""" I want to create a column to indicate my watchlist.  
In this column I would like to have all the series with average IMDb rating above 9.2
indicated as 'To Watch', while the rest will be indicated as 'To Skip'.
"""
#select the series that follow the condition
df.IMDb > 9.2

#now is the time to actually select the rows
df.loc[df.IMDb>9.2]

#assign the value "to watch" for those rows into a new column @watchlist
df.loc[df['IMDb']>9.2, 'watchlist'] ='To Watch'

#assign the value "skip" for the rest 
df.loc[df.IMDb<=9.2, 'watchlist'] ='To Skip'

df.head()
```
