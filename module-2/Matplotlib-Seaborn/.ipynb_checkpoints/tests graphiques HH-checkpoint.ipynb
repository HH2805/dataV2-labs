# Les graphiques dans Matplotlib et Seaborn

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set()
```

```python
x1 = np.arange(0,10,1)
x1
```

```python
x2 = np.linspace(0,5,10)
x2
```

```python
plt.plot(x1, x1, label='linear')
plt.plot(x1, x2**2, label='quadratic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Linear vs Non Linear Plots")
plt.legend()
plt.show()
```

```python
fig, [ax1, ax2] = plt.subplots(1,2)
ax1.plot(x1,x1)
ax1.set_title('LINEAR PLOT')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')

ax2.plot(x1,2**x1)
ax2.set_title('OTHER PLOT')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')

plt.show()
```

```python
url = 'https://studio.ironhack.school/asset-v1:IRONHACK+DAFT+202006_PAR+type@asset+block@datasets_674388_1186156_tv_shows.csv'
df = pd.read_csv(url, index_col='Unnamed: 0')
df.sort_values(by='IMDb', inplace=True)
```

```python
dfsmall = df.iloc[:30,:6]
dfsmall.info()
dfsmall
```

```python
dfsmall = df.iloc[:30,:6]
dfsmall['Genre'] = ['Horror','Romance','Horror','Documentary','Family','Cartoon','Horror','Romance','Documentary','Family','Horror','Romance','Documentary','Family','Cartoon','Horror','Romance','Documentary','Family','Cartoon', 'Horror','Romance','Documentary','Family','Cartoon','Horror','Romance','Documentary','Family','Cartoon']
dfsmall.info()
dfsmall
```

```python
sns.histplot(dfsmall.IMDb)
```

```python
sns.displot(dfsmall.IMDb)
```

```python
dfsmall.boxplot(column = 'IMDb')
```

```python
sns.boxplot(data = dfsmall,
           x = dfsmall.IMDb)
```

```python
sns.violinplot(data = dfsmall, x=dfsmall['IMDb'])
```

```python
sns.countplot(data = dfsmall, x='Genre')
```

```python
sns.lineplot(data = dfsmall, x='Year', y='Netflix')
```

```python
sns.scatterplot(data = dfsmall, x=dfsmall.Year, y=dfsmall.IMDb)
```

```python
ax = sns.barplot(data = dfsmall, x="Genre", y="IMDb")
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

```python
sns.catplot(data = dfsmall, x='Genre', y='IMDb')
```

```python
sns.jointplot(data = dfsmall, x='Year', y='IMDb', kind = 'hex', dropna=True)
```

```python
sns.scatterplot(data = dfsmall, x='Year', y='IMDb', hue='Age')
```

```python
sns.relplot(data = dfsmall, x='Year', y='IMDb', hue='Genre')
```
