```python
import pandas as pd
import numpy as np
```

```python
ees = pd.read_csv('../../../_LabData/employees2.csv')
print(ees.shape)
```

```python
ees['Salary'].value_counts()
```

```python
ees['Salary'].nunique()
```

```python
ees.groupby(['Salary'])
```

```python
ees.groupby(['Department']).mean()
```

```python
ees.groupby(['Department']).mean()['Salary']
```

```python
ees.groupby('Department').median()[['Salary', 'Seniority']]
```

```python
def maxminrange(x):
    return (x.max() - x.min())*100/x.min()
ees.groupby('Department').agg(maxminrange)[['Seniority','Salary']]
```

```python
ees.pivot_table(index=['Department'])
```

```python
ees.pivot_table(index=['Department'], values = ['Salary'])
```

```python
ees.pivot_table(index=['Department'], columns = ['Salary'])
```

```python
ees.pivot_table(index=['Department'], values = ['Salary'], columns = ["Education"])
```

```python
ees.pivot_table(index=["Education",'Department'], values = ['Salary'])
```

```python
ees.pivot_table(index=['Department'], aggfunc={'Salary':['mean','median'],'Age':'count'})
```

```python
ees.pivot_table(index=['Department'], aggfunc={'Salary':['mean','median'],'Age':'count'}, columns = 'Gender')
```

```python
ees.pivot_table(index=['Title','Gender'], aggfunc={'Salary':'median','Age':'median'}, fill_value = 'No Employee')
```

```python
mydict={'Ville':['Paris','Paris','Londres','New York','Londres'],'Temperature':[10,12,12,40,21]}
df = pd.DataFrame(mydict)
```

```python
df.pivot_table(index='Ville', values='Temperature', aggfunc=['count',np.mean,'median','min'])
```

```python
df.pivot_table(index='Ville', aggfunc={'Temperature':['min',np.mean,'median','count']})
```
